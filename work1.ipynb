テストの自分の点数と平均点を入れると推移を示すプログラム

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = ['第1回','第2回','第3回','第4回','第5回']
a,b,c,d,e = int(input('第1回の点数:')),int(input('第2回の点数:')),int(input('第3回の点数:')),int(input('第4回の点数:')),int(input('第5回の点数:'))
y = np.array([[a],[b],[c],[d],[e]])
z = ['自分の点数']
df = pd.DataFrame(data=y,index=x,columns=z)

h,i,j,k,l = int(input('第1回の平均:')),int(input('第2回の平均:')),int(input('第3回の平均:')),int(input('第4回の平均:')),int(input('第5回の平均:'))
y2 = np.array([[h],[i],[j],[k],[l]])
z2 = ['平均']
df1 = pd.DataFrame(data=y2,index=x,columns=z2)
df = pd.concat([df,df1])

idmaxscore = df['自分の点数'].idxmax()
print('平均点：'+str({df['自分の点数'].mean()}))

plt.plot(df['自分の点数'])
plt.plot(df['平均'])
plt.show()
```
